# Analysis of Factors/Modifications Overlapping w/ GR

I first looked at the intersection between each of the factors and GR via venn diagram plotting. After creating and running a script to calculate the proportion of GR that overlaps with each factor, I found that GR has the greatest overlap with H3K4me1 (0.80). In order to evaluate the statistical significance of this overlap, I conducted a Fisher's exact test on the ratio of overlapping to total GR for both H3K4me1 and BCL3 factors. Since the p-value that resulted was less than 0.05, the proportion of GR overlap for H3K4me1 is significantly higher than that for BCL3. Thus, H3K4me1 may closely bind with GR. This is not suprising because H3K4me1 histone modification is a marker for enhancers, which in turn increases gene expression. GR, like H3K4me1, is also known to bind to enhancers. 

```python
from pybedtools import BedTool
import pybedtools
import matplotlib
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy
import math

gr= BedTool("/Users/rzhuang/Documents/ReddyWork/bedtools/gr.bed")
degup= BedTool("/Users/rzhuang/Documents/ReddyWork/bedtools/DEGs_upreg.FDR_0.01.TSS.protein_coding.bed")
nodeg= BedTool("/Users/rzhuang/Documents/ReddyWork/bedtools/non_DEGs.FDR_0.2.TSS.protein_coding.bed")

deggr= str(degup.closest(gr, d= True)) #string of closest features in GR binding sequence to DEG upregulated

#dists = []
#for record in strconv.split("\n"):
    #dists.append(record.split("\t")[-1])
    
dists = [int(record.split("\t")[-1]) for record in deggr.split("\n") if record != '' and '-' not in record]

ndeggr= str(nodeg.closest(gr, d= True)) #string of closest features in GR Binding sequence to Non DEG
ndists= [int(record.split("\t")[-1]) for record in ndeggr.split("\n") if record != '' and '-' not in record]

#bins= numpy.linspace(0, 10000000, num= 25, endpoint=False)
plt.hist(dists, bins=25, alpha= 0.5, label= 'DEG upregulated')
plt.hist(ndists, bins=25, alpha= 0.5, label= 'non DEG')
plt.legend(loc='upper right')
plt.xlabel("Distance of closest GR binding sequence to gene site")
plt.ylabel("Frequency of Distances")
plt.show()
#Mann-Whitney U Test of Distribution

utest=stats.mannwhitneyu(dists, ndists, use_continuity= True, alternative= 'two-sided')
print utest


logdist= [math.log10(distance+1) for distance in dists]
logndist= [math.log10(distance+1) for distance in ndists]
#bins= numpy.linspace(0, 10, num= 25, endpoint=False)
plt.hist(logdist, bins=25, alpha= 0.5, label= 'DEG upregulated', normed=True)
plt.hist(logndist, bins=25 , alpha= 0.5, label= 'non DEG', normed=True)
plt.legend(loc='upper left')
plt.xlabel("Log10 of distance of closest GR binding sequence to gene site")
plt.ylabel("Frequency of distances")
plt.show()
#Mann-Whitney U Test of Distribution

utest=stats.mannwhitneyu(logdist, logndist, use_continuity= True, alternative= 'two-sided')
print utest
```

# Analysis of Distance between GR and DEG Upregulated/Non DEG upon DEX Exposure

I first looked at the distance from each upregulated DEG site/Non DEG site to its closest GR binding site. After computing a list of distances for upregulated DEG site to GR and a list for Non DEG site to GR, I plotted a histogram of the two lists of distances. Since the histogram showed skewed results that were difficult to read, I created another histogram that instead took the log10 of each distance and normalized the distribution generated from each list of distances. Subsequently, I conducted a two-sided Mann Whitney U test on the two lists of distances for Non DEG vs. GR and Upregulated DEG vs. GR. The test generated a P-value substantially smaller than 0.05, supporting the claim that DEG Upregulated sites are significantly closer to GR binding sites than Non DEG sites. This suggests that GR may have a different relationship with Upregulated DEG than Non DEG.
